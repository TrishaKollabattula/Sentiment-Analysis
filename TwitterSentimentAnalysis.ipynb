```python
import re  # Regular expressions for text cleaning

cleaned = []  # List to store cleaned text
for text in df['text']:  # Looping through each message in the 'Message' column
    cleaned_text = re.sub(r'[^\w\s]', '', text)  # Removing all characters except words and spaces
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  # Replacing multiple spaces with a single space
    cleaned_data = cleaned_text.strip()  # Stripping leading and trailing whitespace
    cleaned.append(cleaned_data)
```

```python
tokens = [word_tokenize(x) for x in cleaned]
stop = set(stopwords.words('english'))
stop_token = []
for k in range(len(df['text'])):
    p = [i for i in tokens[k] if i not in stop]
    stop_token.append(p)
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
stemmeddata=[]
for message in stop_token:
   st=[ps.stem(word) for word in message]
   stemmeddata.append(st)
```

```python
from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_token=[pos_tag(text) for text in stop_token]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
X=cv.fit_transform(stem_vector).toarray()
```

```python
y=df['sentiment']
```
